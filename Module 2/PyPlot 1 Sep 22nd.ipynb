```python
# data 특성 보기 by graph
power_data.boxplot(figsize=(17,10))
plt.xticks(rotation=45)
```

```python
power_data.boxplot(['서비스업','제조업'], figsize=(4,8))
plt.xticks(rotation=45)
```

```python
power_data.plot(kind='barh', figsize=(15,8), stacked=True)
```

```python
sample = ['주거용','공공용','서비스업','농림어업','제조업']
power_data[sample].plot(kind='barh', stacked=True, figsize=(12,8))
```

```python
power_data['제조업'].hist()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(power_data['서비스업'], power_data['제조업'])
plt.xlabel('서비스업')
plt.ylabel('제조업')

for n in range(power_data.shape[0]):
    plt.text(power_data['서비스업'][n]*1.02, power_data['제조업'][n]*0.98, power_data.index[n])
plt.show()
```

```python
power = power_data.drop(['서울','경기'])
plt.figure(figsize=(8,8))
plt.scatter(power['서비스업'], power['제조업'])
plt.xlabel('서비스업')
plt.ylabel('제조업')

for n in range(power.shape[0]):
    plt.text(power['서비스업'][n]*1.02, power['제조업'][n]*0.98, power.index[n])
plt.show()
```
